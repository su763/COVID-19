```python


# Read the CSV file
data = pd.read_csv("healthcare_dataset.csv")

# Display the first few rows of the dataframe
display(data.head())
```

```python
# Step 2: Explore the Dataset
print("Dataset Shape:", data.shape)
print("\nFirst 5 Rows:\n", data.head())
print("\nColumn Names:\n", data.columns)
```

```python
# Step 3: Handle Missing Values (if any)
data = data.dropna()  # or use data.fillna(data.mean()) for numerical columns
```

```python

# Step 4: Encode Categorical Columns (if present)
for col in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
```

```python
# Step 3: Handle Missing Values (if any)
# If you chose to drop rows in the previous step, this might not show any missing values.
# If you chose to fill missing values, this will show the remaining missing values.
print("Missing values after handling:")
print(data.isnull().sum())
```

```python
data=data.fillna(data.mean())
```

```python
# Step 5: Split into Features and Target
X = data.drop(columns=['Test Results'])
y = data['Test Results']
```

```python
# Step 6: Scale the Features (if necessary)
# It's often beneficial to scale numerical features for neural networks
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 7: Split the Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

```python
# Step 8: Initialize and Train the MLP Classifier
mlp = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
mlp.fit(X_train, y_train)
```

```python
# Step 9: Make Predictions
y_pred = mlp.predict(X_test)
```

```python
# Step 10: Evaluate Model Performance
print("\n Model Evaluation Results ")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
