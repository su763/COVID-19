```python
# --- 2. Feature Engineering ---
# Convert 'date' column to useful features. The date itself isn't a good predictor.
def create_time_features(df):
    df['date'] = pd.to_datetime(df['date'])
    df['Hour'] = df['date'].dt.hour
    df['DayOfWeek'] = df['date'].dt.dayofweek
    # Drop the original 'date' column
    return df.drop(['date'], axis=1)

train_data = create_time_features(train_data)
test_data = create_time_features(test_data)
```

```python
# --- 3. Define X and y ---
target = 'Occupancy'
features = [col for col in train_data.columns if col != target]

X_train, y_train = train_data[features], train_data[target]
X_test, y_test = test_data[features], test_data[target]
```

```python
# --- 4. Train a Default (Overfit) Tree ---
default_tree = DecisionTreeClassifier(random_state=42)
default_tree.fit(X_train, y_train)

# Evaluate the default tree
y_pred_train_default = default_tree.predict(X_train)
y_pred_test_default = default_tree.predict(X_test)
print("--- Default Tree Evaluation (Task 2) ---")
print(f"Training Accuracy: {accuracy_score(y_train, y_pred_train_default):.4f}")
print(f"Test Accuracy:     {accuracy_score(y_test, y_pred_test_default):.4f}")
print("This large gap shows OVERFITTING.\n")
```

```python
# --- 5. Analyze Feature Importance ---
print("--- Feature Importance ---")
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': default_tree.feature_importances_
}).sort_values(by='Importance', ascending=False)
print(importance_df)
```

```python
# --- 6. Improve Model (Pruning) ---
# Let's train a pruned tree to prevent overfitting by setting max_depth
pruned_tree = DecisionTreeClassifier(max_depth=4, random_state=42)
pruned_tree.fit(X_train, y_train)

# Evaluate the pruned tree
y_pred_train_pruned = pruned_tree.predict(X_train)
y_pred_test_pruned = pruned_tree.predict(X_test)
print("\n--- Pruned Tree Evaluation (max_depth=4) ---")
print(f"Training Accuracy: {accuracy_score(y_train, y_pred_train_pruned):.4f}")
print(f"Test Accuracy:     {accuracy_score(y_test, y_pred_test_pruned):.4f}")
print("The gap is smaller, and test accuracy is higher. The model generalized better.\n")
```

```python
# --- 7. Visualize the Improved Tree ---
print("--- Visualizing the Pruned Tree ---")
plt.figure(figsize=(20, 10))
plot_tree(
    pruned_tree,
    feature_names=features,
    class_names=['Not Occupied', 'Occupied'],
    filled=True,
    rounded=True
)
plt.title("Pruned Decision Tree for Occupancy Detection")
plt.show()
```
